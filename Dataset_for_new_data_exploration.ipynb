```python
import requests
import pandas as pd

# Define the API endpoint for historical daily price data
url = 'https://min-api.cryptocompare.com/data/v2/histoday'

# Define the parameters for the request (e.g., symbol, currency, limit)
params = {
    'fsym': 'BTC',     # From Symbol (Bitcoin)
    'tsym': 'USD',     # To Symbol (US Dollar)
    'limit': 30         # Number of data points to retrieve
}

# Send a GET request to the API endpoint with the parameters
response = requests.get(url, params=params)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the JSON response into a DataFrame
    data = response.json()['Data']['Data']
    df = pd.DataFrame(data)
    print(df.head())
else:
    print(f"Error: {response.status_code}")
```
